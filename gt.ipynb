```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('./gt.csv')
dataset
```

```python
def check_mc(row,value):
  # get all medical conditions and check with input at every row
  all_mc = row['Medical'].split(", ")     
  return value in all_mc

def check_pain(row,value):
  all_pain = row['Pain'].split(", ")
  return value in all_pain

def check_goal(row,value):
   all_goal = row['Goal'].split(", ")
   return value in all_goal
   
input_mc = "Fatigue"
input_pain = "Hips"
input_goal = "Boost Circulation"

# use the check here and store the indexes of mathced rows
mc = dataset[dataset.apply(check_mc,value=input_mc,axis=1)].index
pain = dataset[dataset.apply(check_pain,value=input_pain,axis=1)].index
goal = dataset[dataset.apply(check_goal,value=input_goal,axis=1)].index

#the above datatype is not list so converted in list
mc_yoga=[]
pain_yoga=[]
goal_yoga=[]

for index in mc:
  mc_yoga.append(index)
for index in pain:
  pain_yoga.append(index)
for index in goal:
  goal_yoga.append(index)

print("For Medical do:\t", mc_yoga)
print("For Pain do:\t", pain_yoga)
print("For Goals do:\t", goal_yoga)
```

```python
def check_mc(row, values):
  all_mc = row['Medical'].split(", ")
  for value in values:
    if value in all_mc:
      return True
  return False

def check_pain(row, values):
  all_pain = row['Pain'].split(", ")
  for value in values:
    if value in all_pain:
      return True
  return False

def check_goal(row, values):
  all_goal = row['Goal'].split(", ")
  for value in values:
    if value in all_goal:
      return True
  return False

input_mc = ["Pregnancy", "PCOS"]
input_pain = ["Hips", "Knees"]
input_goal = ["Boost Circulation", "Reduce Stress"]

mc = dataset[dataset.apply(check_mc, values=input_mc, axis=1)]['Name'].tolist()
pain = dataset[dataset.apply(check_pain, values=input_pain, axis=1)]['Name'].tolist()
goal = dataset[dataset.apply(check_goal, values=input_goal, axis=1)]['Name'].tolist()

print("For Medical do:\t", mc)
print("For Pain do:\t", pain)
print("For Goals do:\t", goal)



```

```python
# Define the modified check functions
def check_mc(row, values):
  all_mc = row['Medical'].split(", ")
  score = 0
  for value in values:
    if value in all_mc:
      score += 1
  return score

def check_pain(row, values):
  all_pain = row['Pain'].split(", ")
  score = 0
  for value in values:
    if value in all_pain:
      score += 1
  return score

def check_goal(row, values):
  all_goal = row['Goal'].split(", ")
  score = 0
  for value in values:
    if value in all_goal:
      score += 1
  return score

input_mc = ["Pregnancy", "PCOS"]
input_pain = ["Hips", "Knee"]
input_goal = ["Boost Circulation", "Relaxation"]

# # Get user inputs and split them into lists
# input_mc = input("Enter medical conditions (separated by commas): ")
# input_mc = input_mc.split(", ")
# input_pain = input("Enter pain areas (separated by commas): ")
# input_pain = input_pain.split(", ")
# input_goal = input("Enter goals (separated by commas): ")
# input_goal = input_goal.split(", ")

# Calculate scores for each pose based on the user inputs
mc_scores = dataset.apply(check_mc, values=input_mc, axis=1)
pain_scores = dataset.apply(check_pain, values=input_pain, axis=1)
goal_scores = dataset.apply(check_goal, values=input_goal, axis=1)

# Combine the scores into a matrix
scores = pd.concat([mc_scores, pain_scores, goal_scores], axis=1)
scores.columns = ['Medical', 'Pain', 'Goal']
scores = scores.fillna(0)

# Print the scores matrix
print(scores)

```
